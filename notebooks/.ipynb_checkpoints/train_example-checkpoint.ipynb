# Neural System Identification Training and Prediction Example

This notebook shows the core workflow for one training run and one prediction plot workflow on the IEEE 9-bus dataset using the `wolf_2026_augmented` configuration set.


```python
from copy import deepcopy
import importlib
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import numpy as np
import torch

ROOT = Path.cwd()
if not (ROOT / "neuralsysid").exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from configurations.utils import resolve_base_config, resolve_config_references
from neuralsysid.data import io, preparation, preprocessing
from neuralsysid.training import pipeline
from neuralsysid.utils import plotting
```

## Configuration

The first block controls training. The second block controls the prediction horizon and which rollout window is plotted.


```python
DATA_DIR = ROOT / "data"

DATA_NAME = "2025-10-25_14-57-09_ieee9_kron_"
CONFIG_SET = "wolf_2026_augmented"
MODEL_TYPE = "rk4_ieee9_lstm_noangle"
CHIP_TYPE = "cpu"
NORMALIZE_DATA = True
LOG_MLFLOW = False

EVAL_DATA_NAME = DATA_NAME
EVAL_SPLIT = "test"
PREDICTION_SEQ_LEN = 500
CUTOFF_SAMPLES = 500 - 64
SAMPLE_INDEX = 0
PLOT_GROUP_SIZE = 6
PLOT_X_LIMITS = (0, 500)
PLOT_CHANNEL_SLICE = slice(None)
```

## Train one model

This loads the train/dev/test split, trains the model, and prepares the runtime config and training normalization statistics needed for later prediction plots.


```python
config_module = importlib.import_module(f"configurations.{CONFIG_SET}.base")
model_config = deepcopy(config_module.BASE_CONFIGS[MODEL_TYPE])
model_config["meta"]["config_id"] = f"{CONFIG_SET}:{MODEL_TYPE}:{DATA_NAME}"

train_data_sets, metadata = io.load_data(str(DATA_DIR / DATA_NAME))

model, test_loss = pipeline.learn_dynamics(
    model_config=model_config,
    hyperparameters=[],
    chip_type=CHIP_TYPE,
    data_sets=train_data_sets,
    log_mlf=LOG_MLFLOW,
    normalize_data=NORMALIZE_DATA,
)

runtime_config = resolve_base_config(
    resolve_config_references(model_config, train_data_sets[0][1])
)

train_data_sets_resampled = preprocessing.resample_data_sets(
    train_data_sets,
    t_sample=io.infer_sample_time(train_data_sets[0][0]),
    t_step=model_config["data"]["t_step"],
)
_, data_stats_train = preprocessing.normalize_data(
    train_data_sets_resampled[0][1]
)

print(f"Dataset: {DATA_NAME}")
print(f"Sample time: {io.infer_sample_time(train_data_sets[0][0]):.6f} s")
print(f"Test loss: {test_loss:.6f}")
print(f"Metadata keys: {sorted(metadata)}")
```

## Prepare one evaluation split

This can be the same dataset as above or another local dataset with the same structure. The data is resampled to the model step size and then windowed with a new sequence length for plotting.


```python
eval_data_sets, _ = io.load_data(
    str(DATA_DIR / EVAL_DATA_NAME), extensions=[EVAL_SPLIT]
)
t_eval_raw, data_eval_raw = eval_data_sets[0]

t_eval, data_eval = preprocessing.resample_data_sets(
    [[t_eval_raw, deepcopy(data_eval_raw)]],
    t_sample=io.infer_sample_time(t_eval_raw),
    t_step=runtime_config["t_step"],
)[0]

if NORMALIZE_DATA:
    data_eval_normalized, _ = preprocessing.normalize_data(
        deepcopy(data_eval), data_stats_train
    )
else:
    data_eval_normalized = deepcopy(data_eval)

print(f"Evaluation dataset: {EVAL_DATA_NAME}")
print(f"Evaluation split: {EVAL_SPLIT}")
print(f"Resampled length: {t_eval.shape[1]}")
print(f"Historic sequence length: {runtime_config['hist_seq_len']}")
print(f"Prediction sequence length: {PREDICTION_SEQ_LEN}")
```

## Predict one rollout dataset

This creates a new set of rollout windows with the chosen prediction horizon and then runs the model on all of them. `SAMPLE_INDEX` selects which one you want to inspect.


```python
model.set_n_steps(PREDICTION_SEQ_LEN - 1)

plottable_sets = preparation.generate_batch_sets(
    [t_eval],
    [data_eval_normalized],
    runtime_config["used_outputs"],
    runtime_config["used_controls"],
    batch_size=100000,
    shuffle=False,
    seq_len=PREDICTION_SEQ_LEN,
    historic_seq_len=runtime_config["hist_seq_len"],
    cutoff_samples=CUTOFF_SAMPLES,
)

plottable_sets_true = preparation.generate_batch_sets(
    [t_eval],
    [deepcopy(data_eval)],
    runtime_config["used_outputs"],
    runtime_config["used_controls"],
    batch_size=100000,
    shuffle=False,
    seq_len=PREDICTION_SEQ_LEN,
    historic_seq_len=runtime_config["hist_seq_len"],
    cutoff_samples=CUTOFF_SAMPLES,
)

t_data_eval, _, evaluated_dataloader = plottable_sets[0]
_, _, evaluated_dataloader_true = plottable_sets_true[0]

in_feat, out_feat = list(evaluated_dataloader)[0]
y_hist, u_hist = in_feat[0]
u = in_feat[1]

with torch.no_grad():
    pred_traj = (
        model(y_hist, u_hist, u)[..., : len(runtime_config["used_outputs"])]
        .detach()
        .cpu()
        .numpy()
    )

meas_traj = out_feat.detach().cpu().numpy()
input_traj = u.detach().cpu().numpy()
true_traj = list(evaluated_dataloader_true)[0][1].detach().cpu().numpy()

if NORMALIZE_DATA:
    pred_traj = preprocessing.unnormalize(
        pred_traj,
        data_stats_train["mean_y"][runtime_config["used_outputs"]],
        data_stats_train["std_y"][runtime_config["used_outputs"]],
    )
    meas_traj = preprocessing.unnormalize(
        meas_traj,
        data_stats_train["mean_y"][runtime_config["used_outputs"]],
        data_stats_train["std_y"][runtime_config["used_outputs"]],
    )
    input_traj = preprocessing.unnormalize(
        input_traj,
        data_stats_train["mean_u"][runtime_config["used_controls"]],
        data_stats_train["std_u"][runtime_config["used_controls"]],
    )

print(f"Prediction tensor shape: {pred_traj.shape}")
print(f"Plotting sample: {SAMPLE_INDEX}")
```

## Plot the selected rollout

Use `PLOT_CHANNEL_SLICE` to restrict the shown output channels, for example `slice(0, 3)`.


```python
state_names = [
    data_eval_raw["output_names"][i] for i in runtime_config["used_outputs"]
]
control_names = [
    data_eval_raw["control_names"][i] for i in runtime_config["used_controls"]
]

selected_channels = list(range(len(state_names)))[PLOT_CHANNEL_SLICE]

for start in range(0, len(selected_channels), PLOT_GROUP_SIZE):
    stop = start + PLOT_GROUP_SIZE
    channel_group = selected_channels[start:stop]
    plotting.plot_prediction_comparison(
        measured=meas_traj[SAMPLE_INDEX, :, channel_group],
        reference=true_traj[SAMPLE_INDEX, :, channel_group],
        predicted=pred_traj[SAMPLE_INDEX, :, channel_group],
        controls=input_traj[SAMPLE_INDEX] if start == 0 else None,
        state_names=[state_names[i] for i in channel_group],
        control_names=control_names,
        x_limits=PLOT_X_LIMITS,
    )
    plt.show()
```
